```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open("weather_berlin_2021-06.csv", "r") as csvfile:
    weather_df = pd.read_csv(csvfile, parse_dates=True)
weather_df
```

```python
pv_efficiency = 0.18
pv_area = 1 # m^2
solar = weather_df.solar * pv_area * pv_efficiency
weather_df.solar = solar
ax = weather_df.plot(x='time',y='solar',xlabel='time', ylabel='Power [$W$]', color=['orange'], figsize=(15,4))
ax.figure.savefig('solar_power.png')
```

```python
with open("taxi_2021-06.csv", "r") as csvfile:
    taxi_rides_df = pd.read_csv(csvfile)
taxi_rides_df.head()
```

```python
taxi_rides_df.pickup_datetime = pd.to_datetime(taxi_rides_df.pickup_datetime)
group = taxi_rides_df.groupby(pd.Grouper(freq='1min', key='pickup_datetime'))
taxi_df = group.agg({"dropoff_datetime": "count", 'passenger_count': 'sum'})
taxi_df = taxi_df.rename(columns={"dropoff_datetime": "taxi_count"})
# Saving aggregated dataset to disc
with open("data/taxi_count_2021-06.csv", "w") as csvfile:
    taxi_df.to_csv(csvfile)
taxi_df["2021-06-01 12:00:00":"2021-06-01 12:05:00"]
```

```python
dates = pd.to_datetime(weather_df['time'])
dates.loc[(dates >= "2021-06-01 00:00:00") & (dates <= "2021-06-05 00:00:00")]
```

```python
# Group number of taxis per 15min
taxi_count = taxi_df["taxi_count"]
taxi_count_15min = taxi_count.groupby(pd.Grouper(freq='15min')).sum()

# Plot solar
fig, ax1 = plt.subplots()
ax1.set_ylabel("Watt")
solar_slice = weather_df.loc[(dates >= "2021-06-01 00:00:00") & (dates <= "2021-06-05 00:00:00")]
solar_slice.plot(x='time', y='solar', figsize=(15,4), color="#ff9869", ax=ax1)

# Plot taxis
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel("Taxis per 15min")
taxi_count_slice = taxi_count_15min["2021-06-01 00:00:00":"2021-06-05 00:00:00"]
taxi_count_slice = taxi_count_slice.reset_index()
taxi_count_slice.plot(x='pickup_datetime', y='taxi_count', figsize=(15,4), color="black", ax=ax2)
print(len(solar_slice), len(taxi_count_slice))
# solar_slice.plot(x='time', y='solar')
# taxi_count_15min.reset_index()
```

```python
with open("taxi_2021-06.csv", "r") as csvfile:
    taxi_rides_df = pd.read_csv(csvfile)

taxi_rides_df.pickup_datetime = pd.to_datetime(taxi_rides_df.pickup_datetime)
group = taxi_rides_df.groupby(pd.Grouper(freq='1min', key='pickup_datetime'))
taxi_df = group.agg({"dropoff_datetime": "count", 'passenger_count': 'sum'})
taxi_df = taxi_df.rename(columns={"dropoff_datetime": "taxi_count"})
# Saving aggregated dataset to disc
with open("taxi_count_2021-06.csv", "w") as csvfile:
    taxi_df.to_csv(csvfile)
taxi_df["2021-06-01 12:00:00":"2021-06-01 12:05:00"]
```

```python
# Group number of taxis per 15min
taxi_count = taxi_df["taxi_count"]
taxi_count_15min = taxi_count.groupby(pd.Grouper(freq='15min')).sum()

# Plot solar
fig, ax1 = plt.subplots()
ax1.set_ylabel("Watt")
solar_slice = solar["2021-06-01 00:00:00":"2021-06-05 00:00:00"]
solar_slice.plot.area(figsize=(15,4), color="#EA4444", ax=ax1)

# Plot taxis
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel("Taxis per 15min")
taxi_count_slice = taxi_count_15min["2021-06-01 00:00:00":"2021-06-05 00:00:00"]
taxi_count_slice.plot(figsize=(15,4), color="black", ax=ax2)
fig.savefig('taxi_and_solar.png', bbox_inches='tight', dpi=150)
```
